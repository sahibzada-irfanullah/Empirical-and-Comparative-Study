```python
import pandas as pd
import spacy
import simplejson as sjson
import simplejson as sjson
nlp = spacy.load("en_core_web_lg")
```

```python

path_to_OldData = 'binary_dataset_Copy.json'
j = open(path_to_OldData, encoding="UTF-8")
path_to_NewData = 'binary_dataset_Copy.csv'
df_NewData = pd.read_csv(path_to_NewData, encoding= 'unicode_escape')

count_updated = 0
count_NotUpdatedSimilarFound = 0
count_notSimilarFound = 0
class_updated = 0
df = pd.DataFrame(columns = ["sno", "new_tweet_text", "old_tweet_text", "new_label", "old_label", "updated"], dtype=object)
js_OldData = sjson.loads(j.read())
for _, rowNewData in df_NewData.iterrows():
    print(30*"+")
    similarNotFound = True
    print("New:", rowNewData[0], ":" , rowNewData[1], ":" , rowNewData[2])
    newDataText = nlp(rowNewData[1].lower())
    for i in range(len(js_OldData)):
        OldDataText = nlp(js_OldData[i]['text'].lower())
        
        if newDataText.similarity(OldDataText) > 0.98:
            
            print("SIMILARITY FOUND:")
            print("Old:", js_OldData[i]['text'], ":" , \
                  js_OldData[i]['ugc']['label'])
            if js_OldData[i]['ugc']['label'].replace("not_excl", "not_request") != rowNewData[2]:
                class_updated = 1
                count_updated += 1
            js_OldData[i]['new_text']  = rowNewData[1]
            js_OldData[i]['ugc']['new_label'] = rowNewData[2]
            new_row = {"sno":rowNewData[0], "new_tweet_text":rowNewData[1], "old_tweet_text": \
                       js_OldData[i]['text'], "new_label":rowNewData[2],\
                       "old_label":js_OldData[i]['ugc']['label'], "updated": class_updated}
            df = df.append(new_row, ignore_index=True)
            similarNotFound = False
            class_updated = 0
            break
    if similarNotFound:
        new_row = {"sno":rowNewData[0], "new_tweet_text":rowNewData[1], "old_tweet_text": \
                "Not Found", "new_label":rowNewData[2],\
                "old_label":"Not Found", "updated": 0}
        df = df.append(new_row, ignore_index=True)
        count_notSimilarFound += 1
with open(path_to_OldData + "new11.json", 'w') as fp:
    sjson.dump(js_OldData, fp)  

df.to_csv(path_to_NewData + "details11.csv")
print("Total Number of tweets:", len(js_OldData),\
      ", Number of tweets updated:", count_updated, ", Number of tweets not updated(similar found):", count_NotUpdatedSimilarFound\
     , ", Number of tweets not similar found:", count_notSimilarFound) 
```

```python
import json

path = 'binary_dataset.json'
geo_coord_freq = dict({'Req_geo':0, 'notReq_geo':0, 'Req_coord':0, 'notReq_coord':0}) 

with open(path, 'r') as j:
    data = json.loads(j.read())
    print(len(data))
    for i in range(len(data)):
        label = data[i]['ugc']['label']
        if label == 'request':
            geo_loc = data[i]['geo']
            coord_loc = data[i]['coordinates']
            if geo_loc != None:
                geo_coord_freq['Req_geo'] = geo_coord_freq['Req_geo'] + 1
            if coord_loc != None:
                geo_coord_freq['Req_coord'] = geo_coord_freq['Req_coord'] + 1
        if label == "not_excl":
            geo_loc = data[i]['geo']
            coord_loc = data[i]['coordinates']
            if geo_loc != None:
                geo_coord_freq['notReq_geo'] = geo_coord_freq['notReq_geo'] + 1
            if coord_loc != None:
                geo_coord_freq['notReq_coord'] = geo_coord_freq['notReq_coord'] + 1
print(geo_coord_freq)
```

```python
from collections import Counter

# Open the file in read mode 
text = "this is my file"
  
# Create an empty dictionary 
d = dict() 
# Loop through each line of the file 
for line in text: 
    # Remove the leading spaces and newline character 
    line = line.strip() 
  
    # Convert the characters in line to  
    # lowercase to avoid case mismatch 
    line = line.lower() 
  
    # Split the line into words 
    words = line.split(" ") 
  
    # Iterate over each word in line 
    for word in words: 
        # Check if the word is already in dictionary 
        if word in d: 
            # Increment count of word by 1 
            d[word] = d[word] + 1
        else: 
            # Add the word to dictionary with count 1 
            d[word] = 1
  
c = Counter(d)
```
